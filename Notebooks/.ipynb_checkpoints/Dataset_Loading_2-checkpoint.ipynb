## SARS C3 Protease Assays

```python
data = pd.read_csv('../DatasetAIDs/SARS_C3_Assays_AID_only.txt')
AIDs = list(data.values.squeeze())
assays = []
for AID in AIDs:
    os.system(f'wget https://pubchem.ncbi.nlm.nih.gov/rest/pug/assay/aid/{AID}/csv -O Assay{AID}.csv')
    if os.stat(f'Assay{AID}.csv').st_size != 0:
        assays.append(pd.read_csv(f'Assay{AID}.csv'))
```

```python
for assay in assays:
    if len(assay) != 1:
        cids = list(assay[['PUBCHEM_CID']].values.astype("int32").squeeze())
        nan_counter = 0
        for i in range(len(cids)):
            if cids[i] < 0:
                nan_counter += 1
        cids = cids[nan_counter:]
        mols = []
        for CID in cids:
            #os.system('curl https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/cid/%s/sdf -o cmp.sdf' %CID)
            os.system('wget https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/cid/%s/sdf -O cmp.sdf' %CID)
            time.sleep(.01)
            mols.append(Chem.SDMolSupplier("cmp.sdf")[0])
            
        for i in range(nan_counter):
            mols.insert(0,None)
        
        assay.insert(3, "Mol Object", mols)
        
```

```python
pickle.dump(assays, open(f"/u/macdougt/Research/2019-nCov/Data/sars/sars_assays.pkl", "wb"))
```

```python
assays = {}
assays["sars"] = sars
assays["mers"] = mers
assays["ns3"] = ns3
assays["hiv"] = hiv
```

```python
for a in ["sars", "mers", "ns3", "hiv"]:
    print(len(assays[a]))
    assays[a] = np.array(assays[a])
    bad_list = []
    good_list = []
    for i in range(len(assays[a])):
        ic50_cols = [col for col in assays[a][i].columns if 'PubChem Standard Value' in col]
        if not ic50_cols:
            bad_list.append(i)
        else:
            good_list.append(int(i))

    bad_list = np.array(bad_list)
    good_list = np.array(good_list, dtype='int32')

    assays[a] = assays[a][good_list]
    print(len(assays[a]))
```

```python
for a in ["sars", "mers", "ns3", "hiv"]:
    for i in range(len(assays[a])):
        assays[a][i] = assays[a][i][["Mol Object", "PubChem Standard Value", "Standard Type"]]
```

```python
for a in ["sars", "mers", "ns3", "hiv"]:
    for i in range(len(assays[a])):
        print(assays[a][i][["Standard Type"]].values[-1])
```

```python
counter = 0
for a in ["sars", "mers", "ns3", "hiv"]:
    for i in range(len(assays[a])):
        counter += (len(assays[a][i])-3)
print(counter)
```

```python
all_dfs = []
for a in ["sars", "mers", "ns3", "hiv"]:
    for i in range(len(assays[a])):
        if assays[a][i][["Standard Type"]].values[-1][0] in {"IC50", "Ki", "Kd", "IC90"}:
            all_dfs.append(assays[a][i])
```

```python
counter = 0
for i in range(len(all_dfs)):
    counter += (len(all_dfs[i])-3)
print(counter)
```

```python
for i in range(len(all_dfs)):
    all_dfs[i] = all_dfs[i].iloc[4:]
```

```python
final_df = pd.concat(all_dfs)
```
